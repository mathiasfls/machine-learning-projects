```r
library(caret)
library(tidyverse)
library(scales)
library(cluster)
library(factoextra)
library(NbClust)
```

## Pre-process Data

Standardize iris data

```r
#iris----------
#1. get label
iris.label <- iris$Species
#pre-processing standardize 
preProcValues <- preProcess(iris[,3:4], method = c("center", "scale"))
iris.standard <- predict(preProcValues, iris[,3:4])
head(iris.standard)
```

## Try Elbow Method to find the best number of clusters

For Iris dataset, we know the actual labels, but let's pretend we don't know that and see whether elbow method can give us the best number of cluster.

```r
n_clusters <- 1:8
md.iris.standard <- list()
for(i in seq_along(n_clusters)){
    md.iris.standard [[i]]<- kmeans(x = iris.standard,centers = n_clusters[i],nstart = 20)
}
```

We find that the best number of cluster is 3.

```r
#elbow method
k <- map_dbl(md.iris.standard,~ length(.x$size))
pct_explained <- map_dbl(md.iris.standard,~.x$betweenss/.x$totss)
elbow_data <- tibble(k = k, pct_explained = pct_explained)
options(repr.plot.width=5, repr.plot.height=3.5)
ggplot(elbow_data,mapping = aes(x = k, y = pct_explained)) +
    geom_line() +
    geom_point() +
    scale_y_continuous(labels = percent) +
    geom_vline(xintercept = 3, linetype = 2, color = "skyblue3") +
    geom_text(aes(x = 3, y = 0.8,label = "k = 3"),vjust = "bottom",hjust = "right",color = "skyblue3") +
    labs(x = "number of clusters", y = "percentage variable explained")
```

## silhouette plot

```r
best_kmeans_model <- md.iris.standard[[3]]
sil <- silhouette(x = best_kmeans_model$cluster, #cluster result
           dist = daisy(x = as.matrix(iris.standard),metric = "euclidean")) #distance matrix
options(repr.plot.width=8, repr.plot.height=5.6)
plot(sil,col=1:3)
```

## Plot for cluster and attributes

Also, show actual label to commpare our clustering result

```r
cluster_data <- cbind(iris[,1:4],as.factor(best_kmeans_model$cluster),iris.label)
colnames(cluster_data)[5] <- "cluster"
head(cluster_data)
```
